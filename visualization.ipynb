# 時間変化をプロット

```python
rhs = pd.read_csv('MAC_example/history_of_rhs.csv')
rhs.columns = ['time[s]', 'rhs']
time = rhs['time[s]']; RHS = rhs['rhs']

plt.plot(time, RHS)
plt.xlabel('time[s]'); plt.ylabel('error for p')
plt.yscale('log')
plt.ylim(1e-6, 1e-1)
plt.show()

# 物体の近くでは不安定

rhs2 = pd.read_csv('MAC_example/history_of_rhs2.csv')
rhs2.columns = ['time[s]', 'rhs']
time = rhs2['time[s]']; RHS2 = rhs2['rhs']

plt.plot(time, RHS2)
plt.xlabel('time[s]'); plt.ylabel('error for p')
plt.yscale('log')
plt.ylim(1e-6, 1e-2)
plt.show()

# 物体から離れたところではすぐ収束する

coef = pd.read_csv('MAC_example/history_dimless_coef.csv')
coef.columns = ['time[s]', 'cd', 'cl', 'cp1', 'cp2']
time = coef['time[s]']; cd = coef['cd']; cl = coef['cl']; cp1 = coef['cp1']; cp2 = coef['cp2']

plt.plot(time, cd, label='CD'); plt.plot(time, cl, label='CL'); plt.plot(time, cp1, label='Cp1'); plt.plot(time, cp2, label='Cp2')
plt.ylim(-2.5, 2.5)
plt.xlabel('time[s]')
plt.legend()
plt.show()

# なんか不安定だな，どこかの項がよくないのか，c++とfortranの違いなのか
```
